```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("superstore dataset.csv", encoding='latin1')
```

```python
print("Total Profit:", df['Profit'].sum())
```

```python
print("Average Sales:", df['Sales'].mean())
```

# Sales by Category

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.show()
```

# Profit by Category

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum)
plt.title("Total Profit by Category")
plt.show()
```

# Sales by Region

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title("Sales by Region")
plt.show()
```

# Profit by Region

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Profit', data=df, estimator=sum)
plt.title("Profit by Region")
plt.show()
```

# Monthly Sales Trend

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

monthly_sales.plot(marker='o', figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Top 10 Products

```python
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)

top_products.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Products by Sales")
plt.show()
```

# Segment-wise Sales

```python
sns.barplot(x='Segment', y='Sales', data=df, estimator=sum)
plt.title("Sales by Customer Segment")
plt.show()
```

# Ship Mode Distribution

```python
sns.countplot(x='Ship Mode', data=df)
plt.title("Ship Mode Distribution")
plt.xticks(rotation=20)
plt.show()
```

# Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Sales Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```
